## 3. What is a box tuple, and how does it work?
    > Box tuple is an argument which many of the Pillow functions take. This is a two x,y coordinate for the image placed in a tuple. Like (10, 15), will have a x value (width) of 10, and y value (height) of 15.

## 5. What method would you call to get Image object for a 100100 image, excluding the lower-left quarter of it?
    > 



```python
import numpy as np
adoggo_cropped = adoggo.resize((100, 100))
img_arr = np.array(adoggo_cropped)
```

```python
img_arr[50:100,0:50] = (0,0,0)
img = Image.fromarray(img_arr)
img.show()
```
